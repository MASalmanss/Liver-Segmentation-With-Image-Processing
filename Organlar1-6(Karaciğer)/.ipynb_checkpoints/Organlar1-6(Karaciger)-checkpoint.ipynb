```python
import numpy as np
from sklearn.mixture import GaussianMixture
from skimage.morphology import binary_closing, disk
from scipy.ndimage import median_filter
from skimage import exposure
from skimage.filters import threshold_otsu
import nibabel as nib

# === Fonksiyonlar ===

def load_data(image_path, label_path, liver_label):
    """NIfTI formatındaki görüntü ve etiket verilerini yükler.
    - `image_path`: Görüntü dosyasının yolu.
    - `label_path`: Etiket dosyasının yolu.
    - `liver_label`: Karaciğerin etiket değeri.
    """
    image_nii = nib.load(image_path)
    label_nii = nib.load(label_path)
    image_data = image_nii.get_fdata()
    label_data = label_nii.get_fdata()
    return image_data, label_data, liver_label

def crop_image(image, roi_bounds):
    """Belirtilen ROI (ilgi alanı) sınırlarına göre görüntüyü kırpar.
    - `roi_bounds`: ROI sınırlarını belirtir (x_start, x_end, y_start, y_end).
    """
    x_start, x_end, y_start, y_end = roi_bounds
    return image[y_start:y_end, x_start:x_end]

def preprocess_image(image, steps):
    """Görüntüyü verilen ön işleme adımlarına göre işler.
    - `normalize`: Görüntü piksel değerlerini 0-1 arasına ölçekler.
    - `median`: Median filtre ile gürültüyü azaltır.
    - `histogram`: Histogram eşitleme ile kontrastı artırır.
    """
    for step in steps:
        if step == "normalize":
            image = (image - np.min(image)) / (np.max(image) - np.min(image))
        elif step == "median":
            image = median_filter(image, size=3)
        elif step == "histogram":
            image = exposure.equalize_hist(image)
    return image

def adaptive_otsu_thresholding(image):
    """Adaptif Otsu eşikleme uygular.
    - Görüntüyü bloklara ayırarak her bir blok için ayrı eşik hesaplar.
    """
    block_size = 32
    h, w = image.shape
    otsu_mask = np.zeros_like(image, dtype=bool)
    for i in range(0, h, block_size):
        for j in range(0, w, block_size):
            block = image[i:i+block_size, j:j+block_size]
            if block.size > 0:
                thresh = threshold_otsu(block)
                otsu_mask[i:i+block_size, j:j+block_size] = block > thresh
    return otsu_mask

def gmm_segmentation(image, n_components=4, covariance_type='diag'):
    """Gaussian Mixture Model (GMM) ile görüntü segmentasyonu yapar.
    - `n_components`: Segmentasyon için ayrılacak sınıf sayısı.
    """
    pixels = image.flatten().reshape(-1, 1)
    gmm = GaussianMixture(n_components=n_components, covariance_type=covariance_type, random_state=0)
    gmm.fit(pixels)
    segmented = gmm.predict(pixels).reshape(image.shape)
    return segmented

def apply_morphological_op(mask, operation):
    """Maskeye morfolojik işlemler uygular.
    - `closing`: Maskedeki küçük boşlukları kapatır.
    """
    if operation == "closing":
        return binary_closing(mask, disk(3))
    else:
        return mask

def calculate_iou(mask1, mask2):
    """IoU (Intersection over Union) skorunu hesaplar.
    - İki maske arasındaki kesişim ve birleşim oranını döndürür.
    """
    intersection = np.logical_and(mask1, mask2).sum()
    union = np.logical_or(mask1, mask2).sum()
    return intersection / union if union != 0 else 0

# === Yollar ve Sabitler ===

image_path = "amos_img_0001.nii.gz"  # Görüntü dosyası yolu
label_path = "amos_label_0001.nii.gz"  # Etiket dosyası yolu
LIVER_LABEL = 6  # Karaciğer etiket değeri

# === Veriyi Yükleme ===

image_data, label_data, liver_label = load_data(image_path, label_path, LIVER_LABEL)

# === Parametreler ===

roi_bounds = (150, 520, 100, 400)  # Görüntü kırpma sınırları
best_preprocessing_combo = ["normalize", "median", "histogram"]  # En iyi ön işleme kombinasyonu

# === Tüm Dilimlere İşlem ===

iou_scores = []

for slice_index in range(image_data.shape[2]):
    # Dilimi ve etiketini al
    image_slice = image_data[:, :, slice_index]
    label_slice = label_data[:, :, slice_index]
    
    # Görüntü ve maskeyi kırp
    cropped_image = crop_image(image_slice, roi_bounds)
    cropped_mask = crop_image(label_slice == LIVER_LABEL, roi_bounds)
    
    if np.sum(cropped_mask) == 0:  # Karaciğer olmayan dilimleri atla
        continue
    
    # Görüntüyü ön işle
    processed_image = preprocess_image(cropped_image, best_preprocessing_combo)
    
    # Adaptif Otsu
    otsu_mask = adaptive_otsu_thresholding(processed_image)
    
    # GMM Segmentasyonu
    segmented = gmm_segmentation(processed_image)
    segmented_resized = segmented[:cropped_mask.shape[0], :cropped_mask.shape[1]]
    valid_pixels = segmented_resized[cropped_mask.astype(bool)]
    predicted_class = np.argmax(np.bincount(valid_pixels)) if len(valid_pixels) > 0 else 0
    predicted_mask = segmented_resized == predicted_class
    
    # Morfolojik işlemler
    final_mask = apply_morphological_op(predicted_mask, "closing")
    
    # IoU Skorunu hesapla
    iou_score = calculate_iou(final_mask, cropped_mask)
    iou_scores.append(iou_score)
    print(f"Dilim {slice_index}: IoU Skoru = {iou_score:.4f}")

# === Ortalama IoU ===

average_iou = np.mean(iou_scores)
print(f"\nOrtalama IoU: {average_iou:.4f}")
```

```python
import matplotlib.pyplot as plt

# Dilim numarası
slice_index = 75

# Dilim ve etiketleri al
image_slice = image_data[:, :, slice_index]
label_slice = label_data[:, :, slice_index]

# Görüntü ve maskeyi kırp
cropped_image = crop_image(image_slice, roi_bounds)
cropped_mask = crop_image(label_slice == LIVER_LABEL, roi_bounds)

# Görüntüyü ön işle
processed_image = preprocess_image(cropped_image, best_preprocessing_combo)

# Adaptif Otsu
otsu_mask = adaptive_otsu_thresholding(processed_image)

# GMM Segmentasyonu
segmented = gmm_segmentation(processed_image)
segmented_resized = segmented[:cropped_mask.shape[0], :cropped_mask.shape[1]]
valid_pixels = segmented_resized[cropped_mask.astype(bool)]
predicted_class = np.argmax(np.bincount(valid_pixels)) if len(valid_pixels) > 0 else 0
predicted_mask = segmented_resized == predicted_class

# Morfolojik işlemler
final_mask = apply_morphological_op(predicted_mask, "closing")

# Görselleştirme
plt.figure(figsize=(18, 12))

# Kırpılmış görüntü
plt.subplot(2, 3, 1)
plt.imshow(cropped_image, cmap='gray')
plt.title("Kırpılmış Görüntü")
plt.axis("off")

# Ön işlenmiş görüntü
plt.subplot(2, 3, 2)
plt.imshow(processed_image, cmap='gray')
plt.title("Ön İşlenmiş Görüntü")
plt.axis("off")

# Adaptive Otsu Maskesi
plt.subplot(2, 3, 3)
plt.imshow(otsu_mask, cmap='gray')
plt.title("Adaptive Otsu Maskesi")
plt.axis("off")

# GMM Segmentasyonu
plt.subplot(2, 3, 4)
plt.imshow(segmented_resized, cmap='jet')
plt.title("GMM Segmentasyonu")
plt.axis("off")

# Tahmin edilen maske
plt.subplot(2, 3, 5)
plt.imshow(predicted_mask, cmap='gray')
plt.title("Tahmin Edilen Maske")
plt.axis("off")

# Morfolojik işlem sonrası maske
plt.subplot(2, 3, 6)
plt.imshow(final_mask, cmap='gray')
plt.title("Morfolojik İşlem Sonrası Maske")
plt.axis("off")

plt.tight_layout()
plt.show()
```

1. Kırpılmış Görüntü
Görsel: Kırpılmış Görüntü

İşlem:
Görüntü, ROI (Region of Interest) sınırlarına göre kırpıldı.
Amaç:
Tüm görüntü yerine yalnızca karaciğerin bulunabileceği alanı işlemek.
İşlem süresini azaltmak ve gereksiz bölgelerden gelen gürültüyü önlemek.
Alternatif:
Tüm görüntüyü işlemek. Ancak bu, işlem yükünü artırır ve düşük performansa neden olabilir.

2. Ön İşlenmiş Görüntü
Görsel: Ön İşlenmiş Görüntü

İşlem:
Görüntü sırasıyla normalize edildi, medyan filtre uygulandı ve histogram eşitlemesi yapıldı.
Normalize: Piksel değerlerini [0, 1] aralığına getirir.
Medyan Filtre: Gürültüyü (ör. tuz ve karabiber gürültüsü) azaltır ve kenar bilgilerini korur.
Histogram Eşitleme: Kontrastı artırarak karaciğerin dokusal farklılıklarını daha belirgin hale getirir.
Amaç:
Görüntüyü daha homojen hale getirerek, segmentasyon için daha uygun hale getirmek.
Alternatif:
Gaussian filtre yerine medyan filtre kullanıldı çünkü medyan filtre kenarları daha iyi korur.
Histogram eşitleme yerine CLAHE (Contrast Limited Adaptive Histogram Equalization) kullanılabilirdi, ancak daha maliyetlidir.

3. Adaptive Otsu Maskesi
Görsel: Adaptive Otsu Maskesi

İşlem:
Görüntü bloklara ayrıldı ve her blok için Otsu eşikleme uygulandı.
Amaç:
Homojen olmayan aydınlatma durumlarında (ör. farklı yoğunluk seviyeleri) daha hassas bir maske oluşturmak.
Alternatif:
Global Otsu eşikleme kullanılabilirdi. Ancak, bu yöntem lokal yoğunluk farklarını dikkate almaz ve hassasiyet kaybına neden olabilir.

4. GMM Segmentasyonu
Görsel: GMM Segmentasyonu

İşlem:
Gaussian Mixture Model (GMM) kullanılarak görüntü, farklı yoğunluk seviyelerine göre 4 sınıfa ayrıldı.
Karaciğerin bulunduğu sınıf belirlendi.
Amaç:
Görüntüdeki farklı dokusal yoğunlukları (ör. karaciğer, çevre doku) segmentlere ayırmak.
Alternatif:
Hog gibi yöntemler kullanılabilirdi. Ancak GMM, yoğunlukların daha karmaşık dağılımlarını daha iyi modeller.

5. Tahmin Edilen Maske
Görsel: Tahmin Edilen Maske

İşlem:
GMM ile segmentasyon sonrası, karaciğerin olduğu sınıf seçildi ve bu sınıfa ait piksellerden bir maske oluşturuldu.
Amaç:
Segmentasyon çıktısından yalnızca karaciğerin bulunduğu alanı belirlemek.
Alternatif:
Daha fazla sınıf eklenerek daha detaylı bir segmentasyon yapılabilirdi, ancak bu, gereksiz işlem maliyeti oluşturabilir.

6. Morfolojik İşlem Sonrası Maske
Görsel: Morfolojik İşlem Sonrası Maske

İşlem:
Binary Closing işlemi uygulandı. Bu işlem, küçük boşlukları kapatır ve maskeyi daha bütünsel hale getirir.
Amaç:
Segmentasyon sırasında oluşabilecek hataları (ör. küçük delikler) düzeltmek.
Alternatif:
Opening işlemi kullanılabilirdi, ancak bu yöntem, maskeden küçük segmentleri silebilir ve bilgi kaybına yol açabilir.

7. Hough Dönüşümü

Eğer projenin kapsamı yalnızca karaciğer segmentasyonu ve performans ölçümüyse:

Hough Dönüşümüne gerek yok. Ancak proje daha sonra geometrik yapıların analizini de kapsayacak şekilde genişletilirse:

Hough dönüşümünü belirli bir adım olarak eklemek düşünülebilir.

8. Extrem Nokta Tespiti

Projenin amacı karaciğer segmentasyonu ve IoU metriğiyle doğruluk analizi olduğu için şu anda ekstrem nokta tespitine veya Harris köşe tespitine ihtiyaç yok gibi görünüyor.

Neden Gerek Yok?

Segmentasyon Amacı: Segmentasyon işlemi için piksellerin yoğunluk, doku veya sınıflandırma temelli ayrımı yeterli. Ekstrem nokta tespiti bu hedefe doğrudan katkı sağlamıyor.

IoU Analizi: IoU metriği, maskelerin örtüşme oranını analiz ediyor. Ekstrem noktalar bu analizi iyileştirmek için gerekli değil. Şu Anda Kullanılan Yöntemler:

GMM, HoG, CoHoG, ve LBP gibi yöntemler doğrudan piksellerin özelliklerine dayanıyor. Bu yöntemler Harris gibi köşe tespitinden farklı bir mantıkla çalışıyor.
